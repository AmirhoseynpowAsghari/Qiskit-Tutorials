# Pure State 
<br><br>

Pure quantum states can be thought of as definite quantum states. That is, states in which we have complete
knowledge (statistically speaking) of the underlying wavefunction.

```
Definition (Pure state). A pure state is a quantum state whose wavefunction is known with
certainty.

```

Note that complete knowledge of the underlying wavefunction does not give complete knowledge of
measurement statistics. Quantum measurements are always statistical (sometimes called noisy) due to the
underlying postulates of quantum mechanics which we have previously seen. While this measurement noise
or uncertainty is always present, we can introduce a different type of uncertainty by allowing for a statistical
distribution of wavefunctions. This is the concept of mixed states, formally defined below.

<br><br>


# Mixed state 

<br><br>
We begin with the formal definition of a mixed state.
<br>
Definition (Mixed state). A mixed state is a statistical distribution of pure states. We can write a mixed state as$
\{(p_i , |\psi_i\rangle)\}_{i=1}^n$ where the $ p_i $'s form a valid probability distribution $(\sum_{i \in [n]} p_i = 1$) and each $ |\psi_i\rangle $ is a valid (pure) state.




## Example 1: Example of a Mixed State

Suppose you are in the lab, and your lab mate tells you she’ll prepare the $|0⟩$ state for you with probability $ \frac{2}{3} $ and the $|1⟩$ state with probability $ \frac{1}{3} $. This then defines a mixed state which we may write as

$ 
\left\{ \left( \frac{2}{3}, |0⟩ \right), \left( \frac{1}{3}, |1⟩ \right) \right\}. 
$

Note that this is completely different from the state

$ 
\left( \frac{2}{3} \right)|0⟩ + \left( \frac{1}{3} \right)|1⟩  
$

(which is not a valid quantum state because it is not normalized). Even if it was normalized (what is the normalization factor?), the mixed state (6.2) is still completely different from (6.3). Why is this?

Equation 6.2 refers to the situation in which we have the state $|0⟩$ with probability $ \frac{2}{3} $, and we have the state $|1⟩$ with probability $ \frac{1}{3} $. There is no situation in which we ever have a superposition of $|0⟩$ and $|1⟩$. In contrast, equation (6.3) refers to the situation in which we definitely have the superposition state $ \left( \frac{2}{3N} \right)|0⟩ + \left( \frac{1}{3N} \right)|1⟩ $ where \(N\) is the normalization factor.

This may seem like a subtle point, but it is crucial to understand. Shortly, we’ll introduce a mathematical tool that makes it easier to see the difference between pure and mixed states. However, before moving on, make sure you understand the difference at the conceptual level.

---




### Mixed State vs. Superposition State

#### Mixed State

A mixed state represents a statistical ensemble of different quantum states, each with a certain probability. In your example, the mixed state is described as:

$
\{(2/3, |0\rangle), (1/3, |1\rangle)\}
$

This means:
- With probability $2/3$, the system is in state $ |0\rangle $.
- With probability $1/3$, the system is in state $ |1\rangle $.

Mathematically, the mixed state is represented by a density matrix $ \rho $:

$
\rho = \frac{2}{3} |0\rangle \langle 0| + \frac{1}{3} |1\rangle \langle 1|
$

For the states $ |0\rangle $ and $ |1\rangle $:

$
|0\rangle \langle 0| = \begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix}, \quad
|1\rangle \langle 1| = \begin{pmatrix}
0 & 0 \\
0 & 1
\end{pmatrix}
$

So, the density matrix $ \rho $ is:

$
\rho = \frac{2}{3} \begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix} + \frac{1}{3} \begin{pmatrix}
0 & 0 \\
0 & 1
\end{pmatrix} = \begin{pmatrix}
\frac{2}{3} & 0 \\
0 & \frac{1}{3}
\end{pmatrix}
$

This matrix represents the probabilistic mixture of the states $ |0\rangle $ and $ |1\rangle $.

#### Superposition State

A superposition state is a linear combination of quantum states. In your example, the state is described as:

$
(2/3) |0\rangle + (1/3) |1\rangle
$

However, this is not a valid quantum state because it is not normalized. A quantum state must satisfy the normalization condition:

$
|a|^2 + |b|^2 = 1
$

For the given superposition state:

$
|(2/3)|^2 + |(1/3)|^2 = \left(\frac{2}{3}\right)^2 + \left(\frac{1}{3}\right)^2 = \frac{4}{9} + \frac{1}{9} = \frac{5}{9}
$

To normalize the state, we need to find the normalization factor $ N $ such that:
$
|aN|^2 + |bN|^2 = 1$

Here, \( a = 2/3 \) and \( b = 1/3 \), so:
$
N = \frac{1}{\sqrt{\left(\frac{2}{3}\right)^2 + \left(\frac{1}{3}\right)^2}} = \sqrt{\frac{9}{5}}$

Thus, the normalized superposition state is:
$
N \left( \frac{2}{3} |0\rangle + \frac{1}{3} |1\rangle \right) = \sqrt{\frac{9}{5}} \left( \frac{2}{3} |0\rangle + \frac{1}{3} |1\rangle \right) = \sqrt{\frac{4}{5}} |0\rangle + \sqrt{\frac{1}{5}} |1\rangle$

### Conceptual Difference

1. **Mixed State :**
   - Represents a probabilistic mixture of states.
   - The system is in state $ |0\rangle $ with probability $2/3$ and in state $ |1\rangle $ with probability $1/3$.
   - There is no coherence between $ |0\rangle $ and $ |1\rangle $.
   - Mathematically represented by a density matrix.

2. **Superposition State :**
   - Represents a coherent superposition of states.
   - The system is in a specific quantum state that is a linear combination of $ |0\rangle $ and $ |1\rangle $.
   - There is coherence between $ |0\rangle $ and $ |1\rangle $.
   - Mathematically represented by a state vector in the Hilbert space.

### Density Matrix Representation

To see the difference more clearly, let's look at the density matrices of both cases.

**Mixed State:**

$
\rho_{\text{mixed}} = \frac{2}{3} |0\rangle \langle 0| + \frac{1}{3} |1\rangle \langle 1| = \begin{pmatrix}
\frac{2}{3} & 0 \\
0 & \frac{1}{3}
\end{pmatrix}
$

**Superposition State:**

$
| \psi_{\text{superposition}} \rangle = \sqrt{\frac{4}{5}} |0\rangle + \sqrt{\frac{1}{5}} |1\rangle
$

The density matrix for this pure state is:

$
\rho_{\text{superposition}} = |\psi_{\text{superposition}} \rangle \langle \psi_{\text{superposition}}| = \begin{pmatrix}
\sqrt{\frac{4}{5}} \\
\sqrt{\frac{1}{5}}
\end{pmatrix} \begin{pmatrix}
\sqrt{\frac{4}{5}} & \sqrt{\frac{1}{5}}
\end{pmatrix} = \begin{pmatrix}
\frac{4}{5} & \sqrt{\frac{4}{25}} \\
\sqrt{\frac{4}{25}} & \frac{1}{5}
\end{pmatrix} = \begin{pmatrix}
\frac{4}{5} & \frac{2}{5} \\
\frac{2}{5} & \frac{1}{5}
\end{pmatrix}
$





### Key Differences

1. **Mixed State:**
   - Probabilistic: The system is in either $ |0\rangle $ or $ |1\rangle $ with certain probabilities, but not in both at the same time.
   - No coherence: There is no phase relationship between $ |0\rangle  $ and  $ |1\rangle  $. You simply have a probabilistic mixture.
   - Imagine having two separate jars, one with 2/3 red balls $|0⟩$ and $1/3$ blue balls $|1⟩$. You randomly pick one ball from the combined jar each time you prepare the system.

2. **Superposition State:**
   - Deterministic: The system is in a single state that is a combination of $ |0\rangle $ and $ |1\rangle $.
   - Coherence: There is a specific phase relationship between $ |0\rangle $ and $ |1\rangle $, meaning the state is more than just a simple mixture.
   - Imagine one ball that is both red and blue (a mix of |0⟩ and |1⟩) at the same time. You always prepare this same mixed-color ball each time.





### Part 1: Proving the Density Operator

Given the mixed state:

$
\{(1 - 2\epsilon, |+\rangle), (\epsilon, |0\rangle), (\epsilon, |1\rangle)\}
$

where $ |+\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}} $.

First, we need to construct the density operator $ \rho $ for this mixed state. The density operator for a mixed state is a weighted sum of the projectors onto each state:

$
\rho = (1 - 2\epsilon) |+\rangle \langle +| + \epsilon |0\rangle \langle 0| + \epsilon |1\rangle \langle 1|
$

Let's compute each term separately.

1. **Projector for $ |+\rangle $:**

$
|+\rangle \langle +| = \left( \frac{|0\rangle + |1\rangle}{\sqrt{2}} \right) \left( \frac{\langle 0| + \langle 1|}{\sqrt{2}} \right)
= \frac{1}{2} (|0\rangle \langle 0| + |0\rangle \langle 1| + |1\rangle \langle 0| + |1\rangle \langle 1|)
$

2. **Projector for $ |0\rangle $:**

$
|0\rangle \langle 0| = \begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix}
$

3. **Projector for $ |1\rangle $:**

$
|1\rangle \langle 1| = \begin{pmatrix}
0 & 0 \\
0 & 1
\end{pmatrix}
$

Now, combine these with the given probabilities:

$
\rho = (1 - 2\epsilon) \left( \frac{1}{2} (|0\rangle \langle 0| + |0\rangle \langle 1| + |1\rangle \langle 0| + |1\rangle \langle 1|) \right) + \epsilon |0\rangle \langle 0| + \epsilon |1\rangle \langle 1|
$

Simplify each term:

$
\rho = \frac{1 - 2\epsilon}{2} \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix} + \epsilon \begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix} + \epsilon \begin{pmatrix}
0 & 0 \\
0 & 1
\end{pmatrix}
$

Combine the matrices:

$
\rho = \frac{1 - 2\epsilon}{2} \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix} + \begin{pmatrix}
\epsilon & 0 \\
0 & \epsilon
\end{pmatrix}
$

$
\rho = \frac{1 - 2\epsilon}{2} \begin{pmatrix}
1 & 1 \\
1 & 1
\end{pmatrix} + \begin{pmatrix}
\epsilon & 0 \\
0 & \epsilon
\end{pmatrix}
= \begin{pmatrix}
\frac{1 - 2\epsilon}{2} + \epsilon & \frac{1 - 2\epsilon}{2} \\
\frac{1 - 2\epsilon}{2} & \frac{1 - 2\epsilon}{2} + \epsilon
\end{pmatrix}
$

$
\rho = \begin{pmatrix}
\frac{1 - 2\epsilon}{2} + \epsilon & \frac{1 - 2\epsilon}{2} \\
\frac{1 - 2\epsilon}{2} & \frac{1 - 2\epsilon}{2} + \epsilon
\end{pmatrix}
= \begin{pmatrix}
\frac{1 - 2\epsilon + 2\epsilon}{2} & \frac{1 - 2\epsilon}{2} \\
\frac{1 - 2\epsilon}{2} & \frac{1 - 2\epsilon + 2\epsilon}{2}
\end{pmatrix}
$

$
\rho = \begin{pmatrix}
\frac{1}{2} & \frac{1 - 2\epsilon}{2} \\
\frac{1 - 2\epsilon}{2} & \frac{1}{2}
\end{pmatrix}
$

This matches the given density operator:

$
\rho = \frac{1}{2} \begin{pmatrix}
1 & 1 - 2\epsilon \\
1 - 2\epsilon & 1
\end{pmatrix}
$

### Part 2: Proving the Properties of Mixed States

To prove that this density operator satisfies the four properties of mixed states, let's list the properties and verify them one by one.

#### 1. **Hermitian:**
   - A density matrix $ \rho $ must be Hermitian, which means $\rho = \rho^\dagger $.

$
\rho = \frac{1}{2} \begin{pmatrix}
1 & 1 - 2\epsilon \\
1 - 2\epsilon & 1
\end{pmatrix}
$

Since all the elements are real numbers and $ \rho $ is equal to its transpose conjugate:

$
\rho^\dagger = \rho
$

Thus, $ \rho $ is Hermitian.

#### 2. **Trace is 1:**
   - The trace of a density matrix must be 1.

$
\text{Tr}(\rho) = \frac{1}{2} (1 + 1) = \frac{2}{2} = 1
$

So, the trace is indeed 1.

#### 3. **Positive Semi-Definite:**
   - A density matrix must be positive semi-definite, meaning all its eigenvalues are non-negative.

To check this, we can compute the eigenvalues of $ \rho $. The characteristic polynomial of $ \rho $ is given by:

$
\det(\rho - \lambda I) = 0
$

$
\det\left(\frac{1}{2} \begin{pmatrix}
1 - 2\lambda & 1 - 2\epsilon \\
1 - 2\epsilon & 1 - 2\lambda
\end{pmatrix}\right) = 0
$

$
\left(\frac{1 - 2\lambda}{2}\right)^2 - \left(\frac{1 - 2\epsilon}{2}\right)^2 = 0
$

$
(1 - 2\lambda)^2 = (1 - 2\epsilon)^2
$

$
1 - 2\lambda = \pm (1 - 2\epsilon)
$

This gives us two eigenvalues:

$
\lambda_1 = \epsilon, \quad \lambda_2 = 1 - \epsilon
$

Both $ \epsilon $ and $ 1 - \epsilon $ are non-negative for $ 0 < \epsilon \ll 1 $, so the eigenvalues are non-negative, confirming that $ \rho $ is positive semi-definite.

#### 4. **Unit Trace:**
   - We already verified that the trace of $ \rho $ is 1.

Since $ \rho $ is Hermitian, has a trace of 1, and is positive semi-definite, it satisfies all the properties of a density matrix for a mixed state.
