

### 1. **Schmidt Decomposition and Its Limitation for Fermions**

The **Schmidt decomposition** is a powerful mathematical tool used in quantum mechanics to describe the entanglement in a bipartite quantum system (i.e., a system divided into two subsystems). Given a pure state of two distinguishable subsystems $A$ and $B$, the Schmidt decomposition theorem says that the state can be written as:

$
|\psi\rangle = \sum_i \lambda_i |a_i\rangle \otimes |b_i\rangle
$

where $ \lambda_i $ are the Schmidt coefficients, and $ |a_i\rangle $ and $ |b_i\rangle $ are orthonormal basis states for the subsystems $A$ and $B$. The number of non-zero Schmidt coefficients $ \lambda_i $ gives the **Schmidt rank**, which quantifies the entanglement between the two subsystems.

However, **fermions** pose a challenge to this framework because:

- **Indistinguishability**: Fermions are identical particles, meaning you cannot say which fermion is in which subsystem without violating the principle of indistinguishability.
- **Antisymmetry**: The total wavefunction of a system of fermions must be antisymmetric under the exchange of two particles, in accordance with the Pauli exclusion principle. This antisymmetry introduces complications when trying to separate the system into two distinguishable subsystems.

Due to these factors, a **Schmidt decomposition** in the traditional sense doesn't apply to systems of fermions.

### 2. **Fermionic Entanglement and Schmidt Rank Analogs**

Despite these challenges, it is still possible to describe the entanglement of fermionic systems, but the approach is different. Instead of using the standard Schmidt decomposition, we need to work with tools that respect the antisymmetry of fermions.

#### a. **Slater Rank and Slater Decomposition**

For fermions, a useful analog to the Schmidt decomposition is the **Slater decomposition**. This is used to describe fermionic states in a way that respects their antisymmetric nature. The Slater decomposition of a fermionic pure state can be thought of as an expansion in terms of **Slater determinants**—a mathematical object that describes a fully antisymmetric state of fermions.

- **Slater Rank**: The number of terms (Slater determinants) in this decomposition is referred to as the **Slater rank**. This rank provides an analog to the Schmidt rank for fermions. A fermionic state that can be written as a single Slater determinant (i.e., a state with Slater rank 1) is considered **unentangled**. If more than one Slater determinant is needed (Slater rank > 1), the state is entangled.

#### b. **Entanglement Measures for Fermions**

In fermionic systems, entanglement measures need to account for particle indistinguishability and antisymmetry. Some of the methods that have been developed for this purpose include:

- **Fermionic Entanglement Entropy**: This is a generalization of the von Neumann entropy, tailored to fermionic systems. It quantifies the entanglement by considering the reduced density matrices for fermionic subsystems.
- **Mode Entanglement**: Instead of dividing the system into two distinguishable particles, one can divide it into two **modes** (e.g., spatial regions or energy levels), which can be occupied by fermions. Entanglement can then be measured based on the correlations between these modes.
  
- **Correlation Functions**: Another approach involves using correlation functions that measure the statistical correlations between fermions in different modes, thus giving an indirect measure of entanglement.




### 1. **Schmidt Decomposition: A Recap**

For a bipartite quantum system (let's call the two subsystems $A$ and $B$), the **Schmidt decomposition** theorem states that any pure state $ |\psi\rangle $ of the combined system can be written in the following form:

$
|\psi\rangle = \sum_i \lambda_i |a_i\rangle \otimes |b_i\rangle
$

Here:
- $ |a_i\rangle $ are orthonormal states of subsystem $A$.
- $ |b_i\rangle $ are orthonormal states of subsystem $B$.
- $ \lambda_i $ are the Schmidt coefficients, which are real, non-negative numbers.

The number of non-zero $ \lambda_i $'s is called the **Schmidt rank** of the state $ |\psi\rangle $.

### 2. **Separable vs. Entangled States**

The key to understanding whether the state $ |\psi\rangle $ is **separable** or **entangled** lies in the number of non-zero Schmidt coefficients.

- **Separable State**: A bipartite pure state is **separable** if it can be written as a product of a state in subsystem $A$ and a state in subsystem $B$. Mathematically, this means that $ |\psi\rangle $ can be written as:

  $
  |\psi\rangle = |a\rangle \otimes |b\rangle
  $

  In the language of the Schmidt decomposition, this would mean that there is **only one non-zero Schmidt coefficient** (i.e., $ \lambda_1 $), and the Schmidt rank is 1. This corresponds to:

  $
  |\psi\rangle = \lambda_1 |a_1\rangle \otimes |b_1\rangle
  $

  In this case, $ |\psi\rangle $ is clearly separable, and there is no entanglement between subsystems $A$ and $B$.

- **Entangled State**: A bipartite pure state is **entangled** if it **cannot** be written as a product of states in subsystems $A$ and $B$. In the Schmidt decomposition, this means that there are **two or more non-zero Schmidt coefficients** (i.e., Schmidt rank > 1). The state can then be written as a superposition of product states:

  $
  |\psi\rangle = \lambda_1 |a_1\rangle \otimes |b_1\rangle + \lambda_2 |a_2\rangle \otimes |b_2\rangle + \cdots
  $

  In this case, $ |\psi\rangle $ is **entangled** because it cannot be factorized into a single product of states from $A$ and $B$.

### 3. **Interpretation: When Is $ |\psi\rangle $ Separable?**

- **Separable State**: If the Schmidt decomposition of $ |\psi\rangle $ has **only one term**, meaning only one non-zero Schmidt coefficient, then the state $ |\psi\rangle $ is separable. In other words, it can be written as a tensor product of a state from subsystem $A$ and a state from subsystem $B$. There is no entanglement between the two subsystems.

- **Entangled State**: If the Schmidt decomposition has **two or more terms**, meaning more than one non-zero Schmidt coefficient, then the state $ |\psi\rangle $ is **entangled**. The presence of multiple terms in the Schmidt decomposition indicates that the subsystems $A$ and $B$ are correlated in a way that cannot be described by a simple product state. This correlation is what we refer to as quantum entanglement.

### 4. **Example**

Suppose you have a bipartite system with subsystems $A$ and $B$, and the state $ |\psi\rangle $ is given by:

$
|\psi\rangle = \frac{1}{\sqrt{2}} \left( |0\rangle_A \otimes |0\rangle_B + |1\rangle_A \otimes |1\rangle_B \right)
$

The Schmidt decomposition of this state has two non-zero Schmidt coefficients (both equal to $ \frac{1}{\sqrt{2}} $). Since there are two terms in the decomposition, the state is **entangled**.

In contrast, if the state were:

$
|\psi\rangle = |0\rangle_A \otimes |0\rangle_B
$

Here, the Schmidt decomposition has only one term with a single non-zero Schmidt coefficient (equal to 1), meaning the state is **separable**.



When dealing with **indistinguishable particles**, such as fermions or bosons, the **Schmidt decomposition** is not directly applicable for analyzing entanglement. This is due to the fact that the Schmidt decomposition relies on partitioning the system into distinguishable subsystems, which is not possible when the particles are fundamentally indistinguishable. In systems of indistinguishable particles, the concepts of entanglement need to be redefined, and different methods are used to quantify and analyze entanglement.

### Why Schmidt Decomposition Fails for Indistinguishable Particles

The Schmidt decomposition works for distinguishable particles because it involves separating the system into distinct subsystems (e.g., subsystem $A$ and subsystem $B$). However, for indistinguishable particles:

1. **Indistinguishability**: The particles do not have well-defined individual identities that can be assigned to distinct subsystems. There is no meaningful way to distinguish one particle from another, even in principle.

2. **Symmetry Constraints**: The wavefunction describing a system of identical fermions or bosons must satisfy symmetry constraints (antisymmetry for fermions and symmetry for bosons) under particle exchange. This makes it impossible to treat the particles as separate, distinguishable entities.

Due to these reasons, the Schmidt decomposition, which is based on dividing the system into distinguishable subsystems, does not work for indistinguishable particles.

### Methods for Analyzing Entanglement in Indistinguishable Particles

Despite the challenges posed by indistinguishability, there are alternative methods that have been developed to analyze and quantify entanglement in systems of indistinguishable particles. These methods take into account the symmetry properties of the particles and the structure of the many-body wavefunction.

#### 1. **Slater Rank for Fermions (Slater Decomposition)**

For fermionic systems, one alternative to the Schmidt decomposition is the **Slater decomposition**. This is a way of expressing the state of a system of fermions in terms of **Slater determinants**.

- A **Slater determinant** is a mathematical object that represents an antisymmetric wavefunction for fermions.
- The **Slater rank** is the number of terms needed in the Slater decomposition to fully represent the state. A single Slater determinant corresponds to a non-entangled fermionic state, while a state that requires more than one Slater determinant is considered **entangled**.

In this context, the Slater rank is analogous to the Schmidt rank but tailored to fermionic systems.

#### 2. **Mode Entanglement**

One powerful approach to analyze entanglement in systems of indistinguishable particles is **mode entanglement**. Instead of thinking about the entanglement between particles, we think about the entanglement between **modes** (such as spatial modes or energy levels) that the particles can occupy. This approach is based on the idea that the particles can occupy different quantum states (modes), and the correlations between these modes can be used to quantify entanglement.

- **Mode Decomposition**: Instead of decomposing the wavefunction into terms associated with distinguishable particles, we decompose it into terms associated with different modes. For example, in a two-mode system, you might look at the occupation of two spatial or momentum modes by indistinguishable particles.
- **Mode Entropy**: You can compute an entanglement measure based on the reduced density matrix for a particular mode. This is similar to the von Neumann entropy, but it applies to the occupation numbers of modes rather than individual particles.

Mode entanglement can be used for both fermions and bosons.

#### 3. **Correlation Functions**

Another approach is to use **correlation functions** to quantify entanglement. Correlation functions measure the statistical correlations between different regions of space or between different modes. These correlations can indicate the presence of entanglement even when the particles themselves are indistinguishable.

- For example, in a fermionic system, you can calculate two-point correlation functions that measure the probability of finding two fermions in particular spatial regions. These correlations reflect the entanglement between different parts of the system.
- Similarly, in bosonic systems, correlations in the occupation numbers of different modes can provide information about entanglement.

#### 4. **Occupation Number Entanglement**

This approach focuses on the occupation number basis, where the state of the system is described in terms of how many particles occupy each mode. For indistinguishable particles, the concept of particle entanglement becomes less meaningful, but entanglement can still be defined in terms of the correlations between different occupation numbers.

- **Fermions**: The Pauli exclusion principle limits occupation to one fermion per mode, so the occupation number can be either 0 or 1. The entanglement can be analyzed by looking at the correlations between the occupation numbers in different modes.
- **Bosons**: For bosons, multiple particles can occupy the same mode, and the entanglement can be quantified by looking at the distribution of occupation numbers across different modes.

#### 5. **Reduced Density Matrix Formalism for Indistinguishable Particles**

The reduced density matrix formalism is a standard approach in quantum mechanics to describe the state of a subsystem. For indistinguishable particles, special care is needed to define a meaningful reduced density matrix that respects the symmetries of the system. This can be done by tracing over certain modes or degrees of freedom while taking into account the indistinguishability of the particles.

Once the reduced density matrix is constructed, entanglement can be quantified using standard measures such as the von Neumann entropy or the purity of the reduced state. However, these measures will reflect the entanglement between modes rather than individual particles.

### Summary of Methods

- **Slater Rank (Fermions)**: Quantifies entanglement by decomposing the fermionic wavefunction into Slater determinants.
- **Mode Entanglement**: Measures the entanglement between different quantum modes rather than particles.
- **Correlation Functions**: Uses statistical correlations between different parts of the system to quantify entanglement.
- **Occupation Number Entanglement**: Analyzes the correlations in the occupation numbers of modes.
- **Reduced Density Matrix**: Constructs a reduced density matrix for the system while respecting particle indistinguishability and uses it to quantify entanglement.

Each of these methods is designed to handle the specific challenges posed by the indistinguishability of particles and can be applied to either fermionic or bosonic systems, depending on the nature of the particles involved.



### **Overview: Slater Rank and Entanglement**

In fermionic systems, the **Slater rank** provides a measure of entanglement. The Slater rank refers to the minimum number of **Slater determinants** required to express the wavefunction of a multi-fermion system.

- **Slater Determinant**: This is a determinant constructed from single-particle wavefunctions (orbitals) that describes an antisymmetric fermionic state. For $N$ fermions, a single Slater determinant corresponds to a state where the fermions are not entangled.
- **Entanglement**: If the wavefunction of the system can be written as a single Slater determinant, the state is **not entangled**. If more than one Slater determinant is required, the state is **entangled**.

### **Step-by-Step Example**

Let's consider a system with two fermions in two spatial modes (e.g., two spatial orbitals). We'll determine whether the state is entangled by calculating the Slater rank.

#### **Step 1: Define the Single-Particle Orbitals**

Assume we have two single-particle wavefunctions (orbitals):

- $ \phi_1(x) $ — First orbital (corresponding to mode 1)
- $ \phi_2(x) $ — Second orbital (corresponding to mode 2)

Here, $ x $ represents the spatial coordinates of the particles. We assume that these orbitals are orthonormal, meaning:

$
\langle \phi_1 | \phi_2 \rangle = 0 \quad \text{and} \quad \langle \phi_1 | \phi_1 \rangle = \langle \phi_2 | \phi_2 \rangle = 1
$

#### **Step 2: Construct the Two-Fermion State**

Let's construct a two-fermion state using a linear combination of Slater determinants. For simplicity, assume that the two fermions occupy some combination of the two orbitals $ \phi_1(x) $ and $ \phi_2(x) $.

##### **Case 1: Non-Entangled State (Slater Rank = 1)**

Consider the following wavefunction:

$
|\psi\rangle = \frac{1}{\sqrt{2}} \left( \phi_1(x_1) \phi_2(x_2) - \phi_2(x_1) \phi_1(x_2) \right)
$

This is the **Slater determinant** for two fermions occupying the orbitals $ \phi_1 $ and $ \phi_2 $. The antisymmetry of the wavefunction ensures that the fermions obey the Pauli exclusion principle.

This wavefunction can be written as a **single** Slater determinant:

$
|\psi\rangle = \frac{1}{\sqrt{2}} \begin{vmatrix}
\phi_1(x_1) & \phi_2(x_1) \\
\phi_1(x_2) & \phi_2(x_2)
\end{vmatrix}
$

Since this wavefunction is represented by a single Slater determinant, the **Slater rank** is 1. Therefore, this state is **not entangled**.

##### **Case 2: Entangled State (Slater Rank > 1)**

Now consider a more complicated wavefunction, which is a superposition of two different Slater determinants:

$
|\psi\rangle = \frac{1}{\sqrt{2}} \left( \text{Slater Determinant 1} + \text{Slater Determinant 2} \right)
$

Let's explicitly construct this superposition:

$
|\psi\rangle = \frac{1}{\sqrt{2}} \left( \frac{1}{\sqrt{2}} \left( \phi_1(x_1) \phi_2(x_2) - \phi_2(x_1) \phi_1(x_2) \right) + \frac{1}{\sqrt{2}} \left( \phi_3(x_1) \phi_4(x_2) - \phi_4(x_1) \phi_3(x_2) \right) \right)
$

Here, $ \phi_3(x) $ and $ \phi_4(x) $ are additional orthonormal orbitals.

This wavefunction is a superposition of two Slater determinants. Therefore, the **Slater rank** is 2, meaning that the state **is entangled**.

#### **Step 3: Conclusion**

- **Slater Rank = 1**: If the wavefunction can be written as a single Slater determinant, the fermions are **not entangled**.
- **Slater Rank > 1**: If the wavefunction requires more than one Slater determinant, the fermions are **entangled**.

In **Case 1**, the wavefunction could be written as a single Slater determinant, so the state was **not entangled**.

In **Case 2**, the wavefunction was a superposition of two Slater determinants, so the state was **entangled** with a Slater rank of 2.

### **General Procedure**

1. **Construct the Wavefunction**: Write the wavefunction for the fermionic system.
2. **Decompose into Slater Determinants**: Express the wavefunction as a sum of Slater determinants.
3. **Count the Slater Determinants**: The number of Slater determinants needed to represent the wavefunction gives the Slater rank.
4. **Determine Entanglement**:
   - If the Slater rank is 1, the state is not entangled.
   - If the Slater rank is greater than 1, the state is entangled.

This approach is a powerful way to determine whether fermionic states are entangled based on their Slater rank.

### 1. **System Setup: Two Electrons on Two Lattice Sites**

Consider two electrons, and assume they can occupy two distinct lattice sites, say site $ A $ and site $ B $. Each electron has two possible spin states: **spin-up** $ |\uparrow\rangle $ and **spin-down** $ |\downarrow\rangle $.

In total, there are four possible single-particle states (one electron on one site with a particular spin):
- $ |A \uparrow\rangle $ (Electron at site $ A $ with spin-up)
- $ |A \downarrow\rangle $ (Electron at site $ A $ with spin-down)
- $ |B \uparrow\rangle $ (Electron at site $ B $ with spin-up)
- $ |B \downarrow\rangle $ (Electron at site $ B $ with spin-down)

These four single-particle states can be used to form the basis for the two-particle state of the system. Since we are dealing with fermions (electrons), the total wavefunction must be antisymmetric with respect to particle exchange.

### 2. **State Representation with a 4x4 Matrix**

The wavefunction for two electrons can be written as a combination of these four single-particle states. Denote the four single-particle states as $ |i\rangle $ with $ i = 1, 2, 3, 4 $, corresponding to the states $ |A \uparrow\rangle $, $ |A \downarrow\rangle $, $ |B \uparrow\rangle $, and $ |B \downarrow\rangle $, respectively.

The state of the two-electron system can then be written as:

$
|\Psi\rangle = \sum_{i,j=1}^{4} \rho_{i,j} |i\rangle \otimes |j\rangle
$

Here:
- $ \rho_{i,j} $ are the coefficients of the state $ |\Psi\rangle $ in this two-electron basis.
- $ \rho $ is a **4x4 matrix** that encodes all the information about the two-electron state.

Since electrons are **fermions**, the wavefunction must be **antisymmetric** under particle exchange. This implies that the matrix $ \rho $ must be **antisymmetric**, meaning $ \rho_{i,j} = -\rho_{j,i} $.

### 3. **Antisymmetric Matrix**

The antisymmetric property of the matrix $ \rho $ reflects the Pauli exclusion principle and the requirement that exchanging two identical fermions should result in a sign change of the wavefunction. For a 4x4 matrix $ \rho $, this antisymmetry means:

- The diagonal elements of $ \rho $ must be zero: $ \rho_{i,i} = 0 $
- The off-diagonal elements satisfy $ \rho_{i,j} = -\rho_{j,i} $

This leaves a total of 6 independent elements in the matrix (since an antisymmetric 4x4 matrix has $ \binom{4}{2} = 6 $ independent components).

### 4. **Normalization Condition**

The normalization condition given is:

$
\text{tr}(\rho^\dagger \rho) = \frac{1}{2}
$

This condition ensures that the total probability of the quantum state is normalized properly. The trace $ \text{tr}(\rho^\dagger \rho) $ represents the sum of the probabilities associated with the state, and the factor of $ \frac{1}{2} $ might be related to the specific conventions used in the normalization of the two-electron state.

### 5. **Physical Interpretation**

This formalism captures the quantum state of a system of two indistinguishable fermions (electrons) on two lattice sites. The antisymmetric matrix $ \rho $ encodes the entanglement and correlations between the electrons, respecting the fermionic nature of the particles.

- **Entanglement**: The off-diagonal elements of the matrix $ \rho $ can represent correlations between the two electrons, indicating entanglement between different states. If the state is separable (i.e., non-entangled), $ \rho $ could potentially be expressed as a product of two vectors (which is generally not possible for an antisymmetric matrix unless the state is highly constrained).
  
- **Fermionic Nature**: The antisymmetry of $ \rho $ ensures that the Pauli exclusion principle is respected, meaning that two electrons cannot occupy the same quantum state.
