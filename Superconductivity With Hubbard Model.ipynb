
### 1. **Basic Concept**

The Hubbard model simplifies the complex interactions of electrons in a solid by focusing on two key aspects:
- **Electron Hopping (Kinetic Energy)**: Electrons can move between neighboring sites on a lattice.
- **Electron-Electron Interaction (Potential Energy)**: When two electrons occupy the same site, they experience a repulsive interaction.

### 2. **Mathematical Formulation**

The Hubbard model is usually written as a Hamiltonian, which is an operator that describes the total energy of the system. For a one-band Hubbard model, the Hamiltonian is:

$
\mathcal{H} = -t \sum_{\langle i,j \rangle,\sigma} \left( c_{i\sigma}^\dagger c_{j\sigma} + \text{H.c.} \right) + U \sum_i n_{i\uparrow} n_{i\downarrow}
$

Let’s break this down:

#### a. **Lattice and Sites**
- The model is defined on a lattice, where each site $i$ can be occupied by an electron.
- Each site can hold up to two electrons, one with spin up $(\uparrow$) and one with spin down $(\downarrow$).

#### b. **Hopping Term $-t$**
- The first term in the Hamiltonian describes the kinetic energy of electrons as they move or "hop" between neighboring sites.
- **$t$** is the hopping amplitude or kinetic energy term.
- **$c_{i\sigma}^\dagger$** is the creation operator, which creates an electron with spin $\sigma$ at site $i$.

- **$c_{j\sigma}$** is the annihilation operator, which removes an electron with spin $\sigma$ from site $j$.

- **$\sum_{\langle i,j \rangle,\sigma}$** indicates the sum over all pairs of neighboring sites $i$ and $j$, and over both spin directions.

The term $c_{i\sigma}^\dagger c_{j\sigma}$ allows an electron to hop from site $j$ to site $i$. The Hermitian conjugate $(\text{H.c.}$) ensures that the reverse process (from $i$ to $j$) is also included.

#### c. **Interaction Term $(U$)**
- The second term describes the potential energy associated with electron-electron interactions.
- **$U$** is the on-site Coulomb repulsion, representing the energy cost when two electrons with opposite spins $(\uparrow\) and \(\downarrow$) occupy the same site.
- **$n_{i\sigma} = c_{i\sigma}^\dagger c_{i\sigma}$** is the number operator, counting the number of electrons with spin $\sigma$ at site $i$.

The term $U \sum_i n_{i\uparrow} n_{i\downarrow}$ adds an energy penalty $U$ whenever both spin-up and spin-down electrons are present at the same site.

### 3. **Key Parameters and Physical Regimes**

The behavior of the Hubbard model is determined by the relative strength of \(t\) and \(U\), leading to different physical regimes:

#### a. **Weak Coupling Regime $(U/t \ll 1$)**
- When the interaction $U$ is much smaller than the hopping $t$, the electrons move freely across the lattice, and the system behaves more like a metal.

#### b. **Strong Coupling Regime $(U/t \gg 1$)**
- When the interaction $U$ is much larger than the hopping $t$, the electrons are strongly localized to avoid the high energy cost of double occupancy. This regime can lead to insulating behavior even at half-filling (one electron per site), known as a Mott insulator.

### 4. **Special Cases and Extensions**

#### a. **Half-Filling**
- At half-filling, each site is occupied by exactly one electron on average. In the strong coupling limit $(U/t \gg 1$), this leads to a Mott insulator where the system is insulating due to electron localization.

#### b. **Magnetism**
- The Hubbard model can describe magnetic phenomena. In the strong coupling limit at half-filling, virtual hopping processes lead to an effective spin-spin interaction, described by the Heisenberg model, which can lead to antiferromagnetic order.

#### c. **Superconductivity**
- Away from half-filling, the Hubbard model can describe superconductivity, especially in the context of high-temperature superconductors. The interaction term $U$ can induce effective attractive interactions between electrons, leading to Cooper pairing and superconductivity.

### 5. **Solving the Hubbard Model**

The Hubbard model is difficult to solve exactly, especially in more than one dimension. Various methods are used to study it:

#### a. **Mean-Field Theory**
- This approach approximates the interactions by treating the effect of all other electrons on a given electron as an average (mean field). This simplifies the problem but may miss important correlations.

#### b. **Numerical Methods**
- **Exact Diagonalization**: Can solve small systems exactly by directly diagonalizing the Hamiltonian.
- **Quantum Monte Carlo (QMC)**: Uses statistical methods to simulate the model and extract physical properties.
- **Dynamical Mean-Field Theory (DMFT)**: Maps the lattice problem to an effective single-site problem, capturing local correlations.

#### c. **Variational Approaches**
- Techniques like the Gutzwiller approximation or variational Monte Carlo can be used to study ground state properties by minimizing the energy with respect to trial wavefunctions.

### 6. **Applications and Relevance**

The Hubbard model is central to understanding:
- **Metal-Insulator Transitions**: How a material can transition from a metal to an insulator due to electron-electron interactions.
- **Magnetism**: The origin of magnetic ordering in materials.
- **Superconductivity**: The pairing mechanisms that lead to superconductivity, especially unconventional superconductors like the cuprates.

To understand Cooper pair formation, we need to focus on the case where the interaction $U$ is small or slightly negative (attractive) instead of repulsive. Here's a step-by-step process:

## We can explain that on each site, when U is negative, we have Cooper pairs, which I can understand as being similar to a qubit?

Yes, we can interpret the situation where \( U \) is negative in the Hubbard model as the formation of Cooper pairs, and we can indeed draw an analogy to qubits. Let me explain how this works.

### The Hubbard Model and Cooper Pairing

The Hubbard model is a simplified model used in condensed matter physics to describe the behavior of electrons in a lattice. The Hamiltonian for the Hubbard model is typically written as:

$
\hat{H} = -t \sum_{\langle i, j \rangle, \sigma} \left( c_{i\sigma}^\dagger c_{j\sigma} + \text{h.c.} \right) + U \sum_{i} \hat{n}_{i\uparrow} \hat{n}_{i\downarrow}
$

where:
- $ c_{i\sigma}^\dagger $ and $ c_{i\sigma} $ are the creation and annihilation operators for an electron with spin $ \sigma $ on site $ i $.
- $ t $ is the hopping parameter, describing the probability amplitude for an electron to move between neighboring sites.
- $ U $ is the on-site interaction energy, describing the energy cost when two electrons occupy the same site.
- $ \hat{n}_{i\sigma} = c_{i\sigma}^\dagger c_{i\sigma} $ is the number operator for electrons with spin $ \sigma $ at site $ i $.

### Negative $ U $ and Cooper Pairing

In the usual Hubbard model, $ U $ is positive, representing repulsive interactions between electrons on the same site. However, if $ U $ is negative, this implies an **attractive interaction** between electrons on the same site. In this scenario, electrons are more likely to pair up on a single site, forming what are known as **Cooper pairs**.

#### Cooper Pairs as a Two-Level System (Qubit Analogy)

A Cooper pair consists of two electrons with opposite spins (one spin-up and one spin-down) occupying the same site. In the context of a single site in the lattice, we can think of the possible states as follows:

- **No electrons on the site**: This can be thought of as the "ground state" or $ |0\rangle $.
- **One electron on the site**: This state can have two possible configurations depending on the spin (up or down).
- **Two electrons on the site** (a Cooper pair): This can be thought of as the "excited state" or $ |1\rangle $.

If we focus only on the states with no electrons and with a Cooper pair, we can effectively describe the system at each site using a two-level system, akin to a qubit:

- **State $ |0\rangle $**: The site is empty (no electrons).
- **State $ |1\rangle $**: The site is occupied by a Cooper pair (two electrons with opposite spins).

Thus, the presence or absence of a Cooper pair on a given site can be represented as a qubit.

### Why is this a Valid Analogy?

- **Binary States**: The analogy works because the site can be in one of two distinct states—either occupied by a Cooper pair or unoccupied. This binary nature mirrors the qubit, which can be in state $ |0\rangle $ or $ |1\rangle $.
- **Quantum Coherence**: Just as a qubit can exist in a superposition of states $ |0\rangle $ and $ |1\rangle $, the quantum state at each site in the lattice can also be in a superposition of having a Cooper pair or not.
- **Entanglement**: Sites in a lattice can become entangled due to interactions, just as qubits in a quantum computer can become entangled.

### Applications and Implications

This qubit analogy is particularly useful in the study of **superconductivity** and **quantum computing**:

- **Superconductivity**: In a superconducting material, Cooper pairs move through the lattice without resistance. Describing these pairs as qubits helps in understanding the quantum mechanical behavior of the superconductor.
- **Quantum Computing**: In certain quantum computing architectures (like superconducting qubits), the qubit states are often implemented using states that resemble the Cooper pair occupation of sites.

### Entanglement in Quantum Mechanics

**Entanglement** is a quantum phenomenon where the quantum states of two or more particles become interdependent, meaning the state of one particle cannot be described independently of the state of the other, even when the particles are separated by large distances.

In a quantum system, if two qubits (or more generally, quantum states) are entangled, the overall state of the system is not just a simple product of the states of each individual qubit but rather a superposition of multiple states that cannot be separated. 

### Entanglement in the Context of the Hubbard Model

In the context of the Hubbard model, entanglement can arise between different sites in the lattice due to interactions that correlate the states of electrons on these sites. Here’s how:

#### 1. **Cooper Pair Formation and Entanglement**

When  $ U  $ is negative, Cooper pairs tend to form on the sites. If we consider two neighboring sites (let’s call them site  $A $ and site $ B $), the state of each site can be either:
- $ |0\rangle_A $ (no Cooper pair on site $ A $) or $ |1\rangle_A $ (Cooper pair on site $ A $).
- $ |0\rangle_B $ (no Cooper pair on site $ B $) or $ |1\rangle_B $ (Cooper pair on site $ B $).

The overall state of the two sites can be described by a superposition of all possible configurations, such as:
$
|\Psi\rangle = \alpha |0\rangle_A |0\rangle_B + \beta |0\rangle_A |1\rangle_B + \gamma |1\rangle_A |0\rangle_B + \delta |1\rangle_A |1\rangle_B
$

However, due to the interactions in the system (like hopping terms or additional coupling), the state of the two sites may become **entangled**. A maximally entangled state for two qubits might look something like:
$
|\Psi\rangle = \frac{1}{\sqrt{2}} \left( |0\rangle_A |1\rangle_B + |1\rangle_A |0\rangle_B \right)
$

In this state, if you measure site $ A $ and find it in state $ |0\rangle_A $, you instantly know that site $ B $ must be in state $ |1\rangle_B $, and vice versa.

#### 2. **Entanglement Across the Lattice**

In a larger lattice, entanglement can extend across many sites, leading to complex correlations between different parts of the system. For example:
- **Pairing across sites**: If the system favors certain configurations due to interactions, the presence of a Cooper pair on one site might be correlated with the absence or presence of a pair on another site.
- **Quantum Superposition**: The entire lattice may exist in a superposition of different configurations where the states of all sites are entangled, meaning the quantum state of the entire lattice cannot be factored into a product of states of individual sites.

#### 3. **Role of Tunneling (Hopping) and Interactions**

The hopping term $ t $ in the Hubbard Hamiltonian allows electrons (and thus Cooper pairs) to move between neighboring sites. This movement introduces correlations between sites. When two sites are coupled by hopping, their quantum states can become entangled. If you consider the case where electrons can tunnel between sites, the total quantum state might reflect the possibility of Cooper pairs existing in superpositions across multiple sites, further enhancing the entanglement.

### Entanglement and Superconductivity

In the context of superconductivity:
- The entanglement of Cooper pairs across the lattice can lead to coherent quantum states that span the entire material, which is crucial for the phenomenon of superconductivity.
- The long-range entanglement of Cooper pairs contributes to the resistance-free flow of current because the Cooper pairs behave coherently as a macroscopic quantum state.

### Measuring Entanglement

In practice, measuring entanglement in such a system involves:
- **Correlation Functions**: Calculating correlation functions between sites can reveal the extent of quantum entanglement. For instance, a non-zero correlation between the occupation numbers of different sites might indicate entanglement.
- **Entanglement Entropy**: The entanglement entropy is another measure used to quantify entanglement. For a subsystem of the lattice, the entanglement entropy measures how much information about the entire system is encoded in the subsystem, reflecting the degree of entanglement between the subsystem and the rest of the lattice.
