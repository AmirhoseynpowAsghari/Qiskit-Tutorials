
### Data Encodings in Quantum Machine Learning

1. **Definition of Encoding:**
   - **Encoding in QML**: The process of "loading" a classical data point $ x \in X $ into a quantum state so that it can be processed by a Quantum Neural Network (QNN).
   - **Classical vs. Quantum**: Unlike classical machine learning, where data is straightforwardly fed into algorithms, quantum machine learning involves converting classical data into quantum states, which presents unique challenges.

2. **Mapping Data to Quantum States:**
   - **Encoding Function**: The encoding function maps a data point $ x $ from the set $ X $ (the space of all possible data points) to an $ n $-qubit quantum state, denoted $ D_n $.
   - **Mathematical Notation**: This encoding can be represented as:
     $
     x \mapsto | \psi(x) \rangle
     $
     where $ | \psi(x) \rangle $ is the quantum state representing the data point $ x $.

3. **Full Wavefunction Encoding:**
   - **Exponential Space Saving**: Many QML methods assume a full wavefunction encoding where $ n = \log_2 N $. This means that a quantum state with $ n $ qubits can represent a data point with $ N $ features.
   - **Example**: If $ N = 2^n $, a quantum state of $ n $ qubits can encode $ N $ classical bits of information.
   - **Advantages**: This encoding provides an exponential saving in the amount of space required to store data. In classical terms, representing $ N $ features would require $ N $ classical bits, but in quantum terms, only $ \log_2 N $ qubits are needed.

4. **Trade-off Between Space and Time:**
   - **Time Complexity**: The trade-off for this exponential space saving is an exponential increase in the time required to prepare the quantum state.
   - **Preparation Time**: Preparing a quantum state that encodes $ N $ features generally takes $ O(2^n) $ time, where $ n = \log_2 N $. This means that as the number of qubits $ n $ increases, the time to prepare the quantum state grows exponentially.
   - **Mathematical Representation**: For \( n \) qubits, the time complexity is $ O(2^n) $. Therefore, for a data point with $ N $ features, where $ n = \log_2 N $, the preparation time is $ O(N) $.



### Visual Representation

To help visualize this:

1. **Classical Data to Quantum State**:
   - Data point $ x $ from set $ X $: $ x \in \mathbb{R}^N $
   - Quantum state $ | \psi(x) \rangle $: Encoded as an $ n $-qubit state where $ n = \log_2 N $

2. **Space vs. Time Trade-off**:
   - **Space**: $ N $ classical bits $ \rightarrow \log_2 N $ qubits (exponential saving in space)
   - **Time**: $ O(2^n) $ for preparing $ n $ -qubit state, where $ n = \log_2 N $ (exponential increase in time)



```python
import matplotlib.pyplot as plt
import numpy as np

# Define N values
N = np.arange(1, 1001)

# Calculate logarithm of N base 2
log2_N = np.log2(N)

# Generate plot
plt.figure(figsize=(8, 6))
plt.plot(N, log2_N, label="log2(N)")
plt.xlabel("N")
plt.ylabel("n = log2(N)") # number of qubits 
plt.title("Logarithm of N (base 2) for N = 1 to 1000")
plt.grid(True)
plt.legend()
plt.show()
```



### Step-by-Step Explanation of Data Encoding

#### 1. Initial State Preparation

- **Initial State $ | \phi \rangle $:**
  - The process begins with an initial quantum state. This is typically the "all-zero" state for $ n $ qubits.
  - Mathematically, this state is denoted as:
    $
    | \phi \rangle = |0\rangle^{\otimes n}
    $
  - This notation means that each of the $ n $ qubits is in the state $ |0\rangle $, resulting in a combined state where all qubits are zero.

#### 2. State Preparation Circuit $ S_x $

- **Unitary Operator $ S_x $:**
  - A state preparation circuit, denoted as $ S_x $, is applied to the initial state $ | \phi \rangle $.
  - $ S_x $ is a unitary operator, meaning it preserves the norm of the quantum state and can be reversed by its conjugate transpose $ S_x^\dagger $.
  - This circuit is composed of a series of one- and two-qubit gates that manipulate the qubits to encode the classical data $ x $.

#### 3. Encoding Function $ E(x) $

- **Encoding Process:**
  - The unitary operation $ S_x $ transforms the initial state $ | \phi \rangle $ into a new quantum state that encodes the data $ x $.
  - The encoded state can be represented as:
    $
    x \mapsto E(x) = S_x | \phi \rangle \langle \phi | S_x^\dagger = | x \rangle \langle x | =: \rho_x
    $
  - Here’s what each term represents:
    - $ S_x | \phi \rangle $: The state resulting from applying $ S_x $ to the initial state $ | \phi \rangle $.
    - $ | \phi \rangle \langle \phi | $: The density matrix representation of the initial state.
    - $ S_x^\dagger $: The adjoint (or conjugate transpose) of $ S_x $.
    - $ \rho_x $: The density matrix of the quantum state that encodes the data $ x $.

### Explanation of $ x $ and $ \rho_x $

- ** $x $:**
  - $ x $ represents a classical data point or feature vector from the dataset.
  - In machine learning, $ x $ is typically a vector of numerical values, where each element corresponds to a feature of the data point.

- **$ \rho_x $:**
  - $ \rho_x $ is the density matrix of the quantum state that encodes the classical data $ x $.
  - The density matrix is a representation of the quantum state in terms of probabilities and coherences between different states. In this context, $ \rho_x $ is used to describe the encoded state resulting from applying the state preparation circuit $ S_x $ to the initial state.

### Desirable Properties of $ S_x $

1. **Polynomial Number of Gates:**
   - The number of gates in the state preparation circuit $ S_x $ should be at most polynomial in the number of qubits $ n $. This ensures that the circuit is computationally feasible.

2. **Bijective Encoding Function:**
   - The encoding function $ E(x) $ should be bijective, meaning each classical data point $ x $ maps to a unique quantum state $ \rho_x $. This property is essential to avoid ambiguity in data representation.

3. **Sub-Polynomial Depth:**
   - For practical applications on Near-term Intermediate-Scale Quantum (NISQ) devices, the depth of the circuit (number of sequential gate operations) should be sub-polynomial to minimize errors due to noise and decoherence.

4. **Hardware Efficiency:**
   - The state preparation circuit $ S_x $ should be hardware efficient, meaning that the one- and two-qubit gates should be realizable with minimal overhead. This includes efficient compilation into the quantum computer’s native gate set and minimizing the use of swap gates to connect disjoint qubits.

### Qubit Encoding Example

- **Motivated by NISQ Limitations:**
  - Due to the limitations of NISQ devices, some authors have considered simpler encoding schemes such as qubit encoding.
  - An example of qubit encoding might be:
    $
    | x_i \rangle = \cos(x_i) |0\rangle + \sin(x_i) |1\rangle
    $
  - Here, each feature $ x_i $ of the data point $ x $ is encoded into a single qubit state. This can be extended to multiple features by taking the tensor product of individual qubit encodings.



```python
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
import matplotlib.pyplot as plt
```

```python
def state_preparation_circuit(x):
    """
    Create a state preparation circuit for encoding the classical data point x.
    """
    n = len(x)
    qc = QuantumCircuit(n)
    
    for i, xi in enumerate(x):
        angle = 2 * np.arctan(xi)  # Convert the feature xi to an angle for rotation
        qc.ry(angle, i)  # Apply Ry rotation to the i-th qubit
    
    return qc
```

```python
# Example classical data point
x = [0.1, 0.5, 0.3]

# Create the state preparation circuit
qc = state_preparation_circuit(x)

# Visualize the circuit
qc.draw('mpl')

```

```python
# Use the Aer simulator to simulate the quantum state
simulator = Aer.get_backend('statevector_simulator')

# Transpile and assemble the quantum circuit
transpiled_qc = transpile(qc, simulator)
qobj = assemble(transpiled_qc)

# Execute the circuit on the statevector simulator
result = simulator.run(qobj).result()

# Get the statevector representing the quantum state
statevector = result.get_statevector()

# Plot the Bloch sphere representation of the quantum state
plot_bloch_multivector(statevector)

```

```python
# Print the statevector to see the quantum state
print("Statevector representing the quantum state:")
print(statevector)
```



### Dense Angle Encoding Definition

#### 1. General Definition

**Given:**
- A feature vector $ x = [x_1, x_2, \ldots, x_N]^T \in \mathbb{R}^N $.

**Encoding Map:**
- The dense angle encoding maps $ x $ to a quantum state $ |x\rangle $.

**State Representation:**
- For $ N $ features, the quantum state is represented as:
  $
  |x\rangle = \bigotimes_{i=1}^{\lceil N/2 \rceil} \left( \cos(\pi x_{2i-1})|0\rangle + e^{2\pi i x_{2i}} \sin(\pi x_{2i-1})|1\rangle \right)
  $
  - Here, $ \lceil N/2 \rceil $ denotes the ceiling function of $ N/2 $, which ensures we have enough qubits to encode all features.

**Explanation:**
- Each qubit encodes two features:
  - The amplitude $ \cos(\pi x_{2i-1}) $ and $ \sin(\pi x_{2i-1}) $ for the $(2i-1)$-th feature.
  - The phase $ e^{2\pi i x_{2i}} $ for the $(2i)$-th feature.

### Step-by-Step for Two-Dimensional Data (Single Qubit)

#### 1. Specific Case for $ x \in \mathbb{R}^2 $

**Given:**
- A two-dimensional feature vector $ x = [x_1, x_2] $.

**Encoding Map:**
- The dense angle encoding maps $ x $ to a single-qubit quantum state.

**State Representation:**
- The quantum state $ |x\rangle $ for a two-dimensional feature vector is:
  $
  |x\rangle = \cos(\pi x_1)|0\rangle + e^{2\pi i x_2} \sin(\pi x_1)|1\rangle
  $

**Explanation:**
- The $|0\rangle$ state is weighted by $ \cos(\pi x_1) $.
- The $|1\rangle$ state is weighted by $ \sin(\pi x_1) $ and has a phase factor $ e^{2\pi i x_2} $.

#### 2. Density Matrix Representation

**Density Matrix:**
- The density matrix $ \rho_x $ for the quantum state $ |x\rangle $ is given by:
  $
  \rho_x = |x\rangle\langle x|
  $;
  $
  \rho_x = \begin{pmatrix}
  \cos^2(\pi x_1) & e^{-2\pi i x_2} \cos(\pi x_1) \sin(\pi x_1) \\
  e^{2\pi i x_2} \cos(\pi x_1) \sin(\pi x_1) & \sin^2(\pi x_1)
  \end{pmatrix}
  $

### Generalizing to Arbitrary Functions

**Observation:**
- The dense angle encoding uses trigonometric functions (sine and cosine) and exponential functions for encoding.

**Generalization:**
- This approach can be extended to other functions, not just trigonometric or exponential. The key idea is to map classical data features to quantum states using some functional transformation.



```python


# Define the dense angle encoding circuit for multiple features
def dense_angle_encoding_circuit(x):
    """
    Create a quantum circuit to encode a feature vector x
    using dense angle encoding.
    """
    num_qubits = len(x) // 2
    qc = QuantumCircuit(num_qubits)
    
    for i in range(0, len(x), 2):
        amplitude_angle = np.pi * x[i]
        phase_angle = 2 * np.pi * x[i+1]
        qc.ry(2 * amplitude_angle, i//2)
        qc.rz(phase_angle, i//2)
    
    return qc

# Example feature vector with 8 features
x = [0.3, 0.7, 0.5, 0.9, 0.2, 0.4, 0.8, 0.6]

# Create the dense angle encoding circuit
qc = dense_angle_encoding_circuit(x)

# Visualize the circuit
qc.draw('mpl')

```

```python
# Use the Aer simulator to simulate the quantum state
simulator = Aer.get_backend('statevector_simulator')

# Transpile and assemble the quantum circuit
transpiled_qc = transpile(qc, simulator)
qobj = assemble(transpiled_qc)

# Execute the circuit on the statevector simulator
result = simulator.run(qobj).result()

# Get the statevector representing the quantum state
statevector = result.get_statevector()

# Plot the Bloch sphere representation of the quantum state
plot_bloch_multivector(statevector)

```


**Wavefunction Encoding:**
- The wavefunction encoding maps a classical vector $ x \in \mathbb{R}^N $ to a quantum state $ |x\rangle $.

**Quantum State Representation:**
- The encoded quantum state $ |x\rangle $ is defined as:
  $
  |x\rangle = \frac{1}{||x||_2} \sum_{i=1}^{N} x_i |i\rangle
  $
  - Here, $ x_i $ is the $ i $-th feature of the vector $ x $.
  - $ ||x||_2 $ is the Euclidean norm (L2 norm) of the vector $ x $, defined as:
    $
    ||x||_2 = \sqrt{\sum_{i=1}^{N} x_i^2}
    $
  - $ |i\rangle $ denotes the computational basis state corresponding to the index $ i $.

**Explanation:**
- The quantum state $ |x\rangle $ is a superposition of computational basis states $ |i\rangle $.
- The coefficient of each basis state \( |i\rangle \) is the corresponding feature $ x_i $ normalized by the Euclidean norm of the vector $ x $.

### Step-by-Step Example with a Two-Dimensional Vector

#### 2. Specific Case for $ x = [x_1, x_2]^T $

**Given:**
- A two-dimensional feature vector $ x = [x_1, x_2]^T $.

**Normalization:**
- Calculate the Euclidean norm of $ x $:
  $
  ||x||_2 = \sqrt{x_1^2 + x_2^2}
  $

**Encoded Quantum State:**
- The wavefunction encoding for $ x $ is:
  $
  |x\rangle = \frac{1}{||x||_2} (x_1 |0\rangle + x_2 |1\rangle)
  $
  - Here, $ |0\rangle $ and $ |1\rangle $ are the computational basis states for a single qubit.

#### 3. Density Matrix Representation

**Density Matrix:**
- The density matrix $ \rho_x $ for the quantum state $ |x\rangle $ is given by:
  $
  \rho_x = |x\rangle \langle x|
  $

**Calculation for Two-Dimensional Vector:**
- Substitute the quantum state $ |x\rangle $:
  $
  |x\rangle = \frac{1}{||x||_2} (x_1 |0\rangle + x_2 |1\rangle)
  $
  - The density matrix \( \rho_x \) is:
    $
    \rho_x = \left( \frac{1}{||x||_2} (x_1 |0\rangle + x_2 |1\rangle) \right) \left( \frac{1}{||x||_2} (\langle 0| x_1 + \langle 1| x_2) \right)
    $
    $
    \rho_x = \frac{1}{||x||_2^2} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \begin{pmatrix} x_1 & x_2 \end{pmatrix}
    $
    $
    \rho_x = \frac{1}{||x||_2^2} \begin{pmatrix} x_1^2 & x_1 x_2 \\ x_1 x_2 & x_2^2 \end{pmatrix}
    $

**Simplified Example:**
- For $ ||x||_2 = 1 $, the density matrix simplifies to:
  $
  \rho_x = \begin{pmatrix} x_1^2 & x_1 x_2 \\ x_1 x_2 & x_2^2 \end{pmatrix}
  $
