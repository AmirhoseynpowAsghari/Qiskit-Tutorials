

### Entropy and Information Content

1. **Entropy $ H(X) $**:
   - Entropy is a measure of uncertainty or randomness in a random variable $ X $. It quantifies the amount of information required to describe the random variable.
   - Mathematically, for a discrete random variable $ X $ with probability distribution $ P(X) $:
     $
     H(X) = -\sum_{x \in \mathcal{X}} P(x) \log P(x)
     $
   - Here, $ \mathcal{X} $ is the set of possible values $ X $ can take.

2. **Joint Entropy $ H(X, Y) $**:
   - Joint entropy measures the uncertainty in a pair of random variables $ (X, Y) $. It quantifies the amount of information required to describe both $ X $ and $ Y $ together.
   - Mathematically, for two discrete random variables $ X $ and $ Y $:
     $
     H(X, Y) = -\sum_{x \in \mathcal{X}, y \in \mathcal{Y}} P(x, y) \log P(x, y)
     $

### Mutual Information

1. **Definition of Mutual Information $ I(X; Y) $**:
   - Mutual information measures the amount of information that one random variable contains about another random variable. It quantifies the reduction in uncertainty about $ X $ given knowledge of $ Y $ (and vice versa).
   - Mathematically:
     $
     I(X; Y) = H(X) + H(Y) - H(X, Y)
     $

### Explanation

- **Sum of Entropies $ H(X) + H(Y) $**:
  - This sum represents the total amount of information required to describe both $ X $ and $ Y $ independently.
  - If $ X $ and $ Y $ are completely independent, $ H(X) + H(Y) $ accurately represents the total uncertainty.

- **Joint Entropy $ H(X, Y) $**:
  - This quantity represents the information needed to describe $ X $ and $ Y $ together, accounting for any dependencies between them.

- **Subtraction to Find Mutual Information**:
  - When $ H(X) $ and $ H(Y) $ are added together, any information common to both $ X $ and $ Y $ (i.e., the information that $ X $ and $ Y $ share) is counted twice.
  - To correct for this double-counting, we subtract the joint entropy $ H(X, Y) $. What remains after this subtraction is the mutual information $ I(X; Y) $, which represents the shared information between $ X $ and $ Y $:
    $
    I(X; Y) = H(X) + H(Y) - H(X, Y)
    $

### Relationship with Conditional Entropy

1. **Conditional Entropy $ H(X|Y) $**:
   - Conditional entropy $ H(X|Y) $ measures the amount of uncertainty remaining in $ X $ when $ Y $ is known.
   - Mathematically:
     $
     H(X|Y) = H(X, Y) - H(Y)
     $

2. **Mutual Information in Terms of Conditional Entropy**:
   - Using the definition of conditional entropy, we can express mutual information as:
     $
     I(X; Y) = H(X) - H(X|Y)
     $
   - This equation shows that the mutual information is the reduction in the uncertainty of $ X $ due to the knowledge of $ Y $.




### Quantum Entangled Qubits

Consider a pair of qubits $A$ and $B$ in the entangled Bell state:
$ |\Psi\rangle = \frac{1}{\sqrt{2}} (|00\rangle + |11\rangle) $
This state means that if qubit $A$ is measured in the computational basis and found to be $ |0\rangle $, then qubit $B$ will also be $ |0\rangle $, and if qubit $A$ is found to be $ |1\rangle $, then qubit $B$ will also be $ |1\rangle $.

### Measurement and Information Content

1. **Single Qubit Entropy $ H(A) $ and $ H(B) $**:
   - Each qubit alone has two possible states, $ |0\rangle $ and $ |1\rangle $, with equal probabilities $ \frac{1}{2} $.
   - The entropy $ H(A) $ for qubit $ A $ is:
     $
     H(A) = -\left( \frac{1}{2} \log \frac{1}{2} + \frac{1}{2} \log \frac{1}{2} \right) = \log 2 = 1 \text{ bit}
     $
   - Similarly, $ H(B) = 1 $ bit.

2. **Joint Entropy $ H(A, B) $**:
   - The joint state of the system is the entangled state $ |\Psi\rangle $.
   - For the entangled state, the outcomes are perfectly correlated:
     $
     P(00) = \frac{1}{2}, \quad P(11) = \frac{1}{2}, \quad P(01) = 0, \quad P(10) = 0
     $
   - The joint entropy $ H(A, B) $ is:
     $
     H(A, B) = -\left( \frac{1}{2} \log \frac{1}{2} + \frac{1}{2} \log \frac{1}{2} \right) = \log 2 = 1 \text{ bit}
     $

3. **Mutual Information $ I(A; B) $**:
   - Mutual information $ I(A; B) $ measures the amount of information shared between $ A $ and $ B $.
   - Using the formula:
     $
     I(A; B) = H(A) + H(B) - H(A, B)
     $
   - Substituting the values:
     $
     I(A; B) = 1 + 1 - 1 = 1 \text{ bit}
     $
   - This indicates that knowing the state of one qubit gives us one bit of information about the state of the other qubit.

4. **Conditional Entropy $ H(A|B) $**:
   - Conditional entropy $ H(A|B) $ measures the remaining uncertainty about $ A $ given that $ B $ is known.
   - For the entangled state:
     $
     H(A|B) = H(A, B) - H(B)
     $
   - Substituting the values:
     $
     H(A|B) = 1 - 1 = 0 \text{ bits}
     $
   - This means there is no uncertainty about $ A $ once $ B $ is known, consistent with perfect correlation in the entangled state.





### Decoherence and Correlations

1. **Decoherence**:
   - Decoherence causes the system to lose coherence, making it behave more classically. It typically results from interactions with the environment, leading to the degradation of quantum properties such as superposition and entanglement.

2. **Quantum vs. Classical Correlations**:
   - Quantum correlations include quantum discord and entanglement.
   - Classical correlations are those that can be explained without quantum mechanics, often involving separable states where each part of the system can be described independently.

### Behavior of Quantum Discord under Decoherence

Researchers have investigated how quantum discord behaves in open quantum systems, which are systems that interact with their environment. One significant finding is that quantum discord can be more robust against decoherence than entanglement, meaning that some quantum correlations may persist even when entanglement is lost.

### Key Findings

1. **Frozen Quantum Discord**:
   - In certain conditions, quantum discord can remain unaffected by decoherence for long periods. This phenomenon, where quantum discord remains constant while classical correlations decay, is often referred to as "frozen discord."

2. **Sudden Transition Between Classical and Quantum Decoherence**:
   - Researchers have identified a sudden transition between two distinct phases:
     - **Classical Decoherence Phase**: In this phase, classical correlations degrade due to decoherence, while quantum discord remains unchanged.
     - **Quantum Decoherence Phase**: In this phase, quantum discord starts to deteriorate, but classical correlations remain constant.

### Example: Two-Qubit System in Markovian Dephasing Channels

1. **Markovian Dephasing**:
   - A Markovian process is one where the system's evolution only depends on its current state, not on its past history. Dephasing channels specifically cause loss of coherence between different quantum states without affecting the populations of the states.

2. **Two-Qubit System**:
   - Consider an open quantum system consisting of two qubits. When these qubits are subject to Markovian dephasing channels, they exhibit the sudden transition behavior:
     - Initially, in the classical decoherence phase, the classical correlations are lost due to the interaction with the environment, but quantum discord remains constant.
     - After a certain point, the system transitions to the quantum decoherence phase, where quantum discord begins to decay, but the classical correlations stabilize.

### Implications

1. **Quantum Information Processing**:
   - The persistence of quantum discord under decoherence suggests that quantum information protocols could still function even when entanglement is not present, as long as quantum discord is maintained.

2. **Quantum Communication and Cryptography**:
   - Understanding the behavior of quantum discord in decoherence can enhance the reliability of quantum communication systems, making them more robust against environmental interactions.

3. **Thermodynamics and Quantum Control**:
   - The ability to manage and control quantum correlations in the presence of decoherence is crucial for the development of quantum thermodynamic systems and quantum control mechanisms.



The generic memory kernel equation is a concept from the theory of open quantum systems. It describes the dynamics of a quantum system interacting with its environment, taking into account the memory effects of the environment. This approach contrasts with Markovian dynamics, where the system's evolution is memoryless and only depends on its current state.

### Memory Kernel Equation

In the study of open quantum systems, the memory kernel equation is a generalized form of the master equation that includes non-Markovian effects. The dynamics of the system are no longer solely determined by its instantaneous state but also by its past states.

#### General Form

The memory kernel equation can be written as:

$ \frac{d}{dt} \rho(t) = \int_0^t K(t - t') \rho(t') dt' $

Here:
- $\rho(t)$ is the density matrix of the system at time $t$.
- $K(t - t')$ is the memory kernel, a function that describes how the past states of the system influence its future evolution.

### Markovian vs. Non-Markovian Dynamics

1. **Markovian Dynamics**:
   - In Markovian dynamics, the system's evolution depends only on its current state, implying no memory effects. The master equation for Markovian dynamics is typically of the Lindblad form:
     $ \frac{d}{dt} \rho(t) = \mathcal{L} \rho(t) $
   - Here, $\mathcal{L}$ is the Lindblad superoperator.

2. **Non-Markovian Dynamics**:
   - Non-Markovian dynamics account for memory effects, meaning the system's evolution depends on its history. The memory kernel equation is used to model such dynamics.

### Properties of the Memory Kernel

- **Non-Locality in Time**: The kernel $K(t - t')$ captures the non-local interactions in time, indicating that the system's future state depends on its entire history.
- **Decay and Oscillations**: The form of $K(t - t')$ can show various behaviors such as exponential decay, oscillations, or more complex forms depending on the nature of the system-environment interaction.

### Applications

The memory kernel approach is used in various areas of quantum physics to model systems where memory effects are significant:

1. **Quantum Optics**: Modeling photon interactions with atoms where the photon's field retains memory of its interactions.
2. **Condensed Matter Physics**: Describing electron-phonon interactions in materials where phonons have a memory of past electron interactions.
3. **Quantum Information**: Understanding decoherence and noise in quantum computing devices where environmental memory plays a crucial role.

### Example: Spin-Boson Model

One common example where memory kernels are used is the spin-boson model. This model describes a two-level system (spin) interacting with a bosonic environment. The dynamics of the spin can exhibit non-Markovian behavior due to the memory effects of the bosonic bath. The memory kernel equation helps to capture this complex interaction.

### Solving the Memory Kernel Equation

Solving the memory kernel equation can be challenging due to its integro-differential nature. Techniques include:

1. **Laplace Transform Methods**: Transforming the equation into the Laplace domain, solving it, and then transforming back.
2. **Perturbative Approaches**: Assuming weak system-environment coupling and using perturbation theory.
3. **Numerical Methods**: Directly integrating the equation numerically for specific forms of the memory kernel.

4. **Spectral Density Method**

### 1. Quantum Mutual Information

#### Definition:
The quantum mutual information$ I(\rho_{AB})$ is defined as:
$
I(\rho_{AB}) = S(\rho_A) + S(\rho_B) - S(\rho_{AB}),$

where:

-$ \rho_{AB}$ is the density matrix of the total quantum system, which includes both subsystems$ A$ and$ B$.
-$ \rho_A$ and$ \rho_B$ are the reduced density matrices of subsystems$ A$ and$ B$, respectively. These are obtained by taking the partial trace over the other subsystem.
-$ S(\rho) = -\text{Tr}(\rho \log_2 \rho)$ is the von Neumann entropy, which is a measure of the uncertainty (or lack of information) in a quantum state$ \rho$.

#### Explanation:
- The quantum mutual information $ I(\rho_{AB}) $ quantifies the total amount of correlations (both quantum and classical) between subsystems $ A $ and $ B $. 
- The term $ S(\rho_A) + S(\rho_B) $ represents the sum of uncertainties of the individual subsystems. 
- The term $ S(\rho_{AB}) $ represents the uncertainty of the combined system.
- The difference between these quantities gives the mutual information, which measures how much knowing the state of subsystem $ A $ reduces the uncertainty about subsystem $ B $, and vice versa.

### 2. Quantum Conditional Entropy

#### Definition:
The second extension involves the concept of **conditional entropy**. This is more complicated because in the quantum setting, the measurement on one subsystem affects the other. Specifically, the quantum conditional entropy $ J(\rho_{AB}|B_k) $ is given by:

$
J(\rho_{AB}|B_k) = S(\rho_A) - S(\rho_{AB}|B_k),
$

where:

- $ B_k $ represents a measurement performed on subsystem $ B $, described by a set of orthonormal projectors $ \{ \Pi_k \} $.
- $ \rho_k $ is the state of the total system after the measurement outcome labeled by $ k $ is obtained. It is given by:

$
\rho_k = \frac{(I \otimes B_k)\rho_{AB}(I \otimes B_k)}{p_k},
$

where $ p_k = \text{Tr}[(I \otimes B_k)\rho_{AB}(I \otimes B_k)] $ is the probability of obtaining outcome $ k $, and $ I $ is the identity operator for subsystem $ A $.

#### Explanation:
- $ J(\rho_{AB}|B_k) $ measures how much the knowledge of subsystem $ B $ (after a specific measurement $ B_k $ is performed) reduces the uncertainty about subsystem $ A $.
- This definition generalizes the classical conditional entropy to the quantum setting. 
- The key idea is that the measurement process in quantum mechanics can affect the state of the system, and this must be accounted for in defining conditional entropy.

### 3. Classical Correlations

#### Definition:
The classical correlations $ C(\rho_{AB}) $ in the state $ \rho_{AB} $ can be defined as:

$
C(\rho_{AB}) = \max_{\{ \Pi_k \}} J(\rho_{AB}|B_k),
$

where the maximization is taken over all possible measurements (sets of projectors $ \{ \Pi_k \} $) on subsystem $ B $.

#### Explanation:
- The idea here is that by finding the measurement on $ B $ that maximizes $ J(\rho_{AB}|B_k) $, we obtain a measure of the classical correlations between $ A $ and $ B $.
- Classical correlations refer to the information that can be extracted from the system through measurements without disturbing it. This is contrasted with quantum correlations, which involve entanglement and other non-classical effects.

### 4. Quantum Discord

- Therefore, the diference between the quantum mutual information $I(\rho_{AB})$,
    describing total correlations, and the measurement-based definition of quantum
    mutual information $C(\rho_{AB})$, measuring classical correlations, defines the so-called
    quantum discord
<br>
$D(\rho_{AB}) = I(\rho_{AB}) - C(\rho_{AB}) $

   - $\chi_\rho^{CD}(t)$ is the closest classical state, indicating that it best approximates the quantum state $\rho(t)$ with no quantum correlations.

2. **At the Point of Transition**:
   - The transition occurs when the weight $\lambda_3$ of the third-largest Bell state component exceeds the weight $\lambda_2$ of the second-largest component.
   - At this critical point (denoted by $\rho_{ST}$ in the figure), two classical states are equally close to the quantum state $\rho(t)$: $\chi_\rho^{CD}(t)$ and $\chi_\rho^{DD}(t)$.

3. **After the Transition**:
   - **$\chi_\rho^{CD}(t)$** no longer remains the closest classical state. The quantum state $\rho(t$) moves away from the trajectory associated with $\chi_\rho^{CD}(t)$.
   - **$\chi_\rho^{DD}(t)$** becomes the new closest classical state. This state lies along the trajectory associated with "decaying discord," meaning that the quantum correlations between $\rho(t)$ and this new classical state begin to decrease.

4. **What Happens to $\chi_\rho^{CD}(t)$**:
   - After the transition, $\chi_\rho^{CD}(t)$ still exists as a classical state, but it no longer represents the best classical approximation of the quantum state $\rho(t)$. In other words, it becomes less relevant in describing the quantum state’s dynamics.
   - The red dotted line representing $\chi_\rho^{CD}(t)$'s trajectory effectively ends or becomes disconnected from the quantum state $\rho(t)$ after the transition.


```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from qiskit.providers.aer.noise import NoiseModel, phase_damping_error

# Create a quantum circuit with one qubit
qc = QuantumCircuit(1, 1)

# Prepare the qubit in a superposition state (|+> state)
qc.h(0)

# Apply the phase damping noise channel (simulates σz decoherence)
prob_dephase = 0.5  # Increase probability to make the effect more pronounced
noise_model = NoiseModel()
phase_damp = phase_damping_error(prob_dephase)
noise_model.add_all_qubit_quantum_error(phase_damp, ['u3', 'h'])

# Apply a Hadamard gate to measure in the X basis (|+> and |->)
qc.h(0)

# Add a measurement
qc.measure(0, 0)

# Execute the circuit on the qasm simulator with the noise model
backend = Aer.get_backend('qasm_simulator')
result = execute(qc, backend, noise_model=noise_model, shots=1000).result()

# Get the counts (measurement results)
counts = result.get_counts(qc)
print("Counts:", counts)

# Plot the result as a histogram
plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Kraus, SuperOp
from qiskit.visualization import plot_histogram
import numpy as np

# Function to create phase flip channel
def phase_flip_channel(p):
    K0 = np.sqrt(1 - p) * np.array([[1, 0], [0, 1]])  # Identity operator
    K1 = np.sqrt(p) * np.array([[1, 0], [0, -1]])    # Sigma_z operator
    return Kraus([K0, K1])

# Example probability of phase flip
p = 0.3

# Initialize a quantum circuit with 1 qubit
qc = QuantumCircuit(1)

# Start with a Hadamard gate to create superposition (|+> state)
qc.h(0)

# Apply the phase flip channel using the Kraus operator representation
phase_flip = phase_flip_channel(p)
qc.append(phase_flip.to_instruction(), [0])

# Draw the circuit
qc.draw('mpl')

# Use the Aer simulator
backend = Aer.get_backend('statevector_simulator')

# Execute the circuit
result = execute(qc, backend).result()

# Get the resulting statevector
statevector = result.get_statevector()
print("Statevector after applying phase flip channel:", statevector)

# Add measurement to the circuit
qc.measure_all()


# Use the Aer qasm simulator for measurement probabilities
qasm_backend = Aer.get_backend('qasm_simulator')

# Execute the circuit with measurements
qasm_result = execute(qc, qasm_backend, shots=1024).result()

# Get the counts (probabilities) of each outcome
counts = qasm_result.get_counts()

# Plot the results
plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from qiskit.providers.aer.noise import NoiseModel, amplitude_damping_error, depolarizing_error, pauli_error, phase_damping_error
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector

# Create a quantum circuit with 3 qubits
qc = QuantumCircuit(3)

# Step 1: Create |phi^+_AB> on qubits 0 (A) and 1 (B)
qc.h(0)            # Apply H gate to qubit 0 (A)
qc.cx(0, 1)        # Apply CNOT with qubit 0 (A) as control and qubit 1 (B) as target

# Step 2: Superposition of qubit C (qubit 2)
qc.h(2)            # Create superposition |+> on qubit 2 (C)

# Step 3: Controlled Z gate to create |phi^-_AB> when C is |-> (conditional phase flip)
qc.cz(2, 0)        # Controlled Z with qubit 2 (C) as control and qubit 0 (A) as target

# Step 4: Apply a CNOT from qubit C (2) to B (1) to entangle C with B
qc.cx(2, 1)        # CNOT to entangle C and B

# View the circuit
qc.draw('mpl')

# Simulate the final state
backend = Aer.get_backend('statevector_simulator')
result = execute(qc, backend).result()

# Get the counts (probabilities) of each outcome
counts = result.get_counts()

# Plot the results
plot_histogram(counts)

statevector = result.get_statevector()

# Plot Bloch vectors and Qsphere
plot_bloch_multivector(statevector)
plot_state_qsphere(statevector)


```
